```python
class Bank:
  bankname="State Bank Of India"    #class variable
  branch="Anand Branch"

  #create account
  def __init__(self,username,accno,address):
    self.username=username
    self.accno=accno
    self.address=address
    self.balance=0

  #deposite
  def deposit(self,amount):
    self.balance=self.balance+amount
    print(f"{amount} deposited successfully")

  #withdraw
  def withdraw(self,amount):
    if amount<self.balance:
      self.balance=self.balance-amount
      print(f"{amount} your withdraw successfully")
    else:
      print("Inssficient fund....")

  #ministatement
  def ministatement(self):
    print(f"your current balance is {self.balance}")

  def display(self):
    print("account holder information")
    print(f"Bank Name:{self.bankname}")
    print(f"user name: {self.username}")
    print(f"account number:{self.accno}")
    print(f"address:{self.address}")
    print(f"balance:{self.balance}")

# collect user data for account creation
print(f"Welcome to {Bank.bankname}, {Bank.branch}")
username=input("enter the username:")
accno=int(input("enter the account no:"))
address=input("enter the address:")

b=Bank(username,accno,address)
while True:
  print("----select any option----")
  print("1. deposit")
  print("2. withdraw")
  print("3. ministatement")
  print("4. Display Information")
  print("5. stop")
  option=int(input("enter the choice:"))
  if option==1:
    amount=float(input("enter the deposited amount:"))
    b.deposit(amount)
  elif option==2:
    amount=float(input("enter the withdarw amount:"))
    b.withdraw(amount)
  elif option==3:
    b.ministatement()
  elif option==4:
    b.display()
  elif option==5:
    print("thank you for using state bank of india")
    break
  else:
    print("invalid option")
```
